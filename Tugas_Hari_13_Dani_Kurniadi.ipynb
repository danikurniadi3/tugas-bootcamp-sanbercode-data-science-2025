<h2>Soal 1: Multiple Subplots</h2>

Dengan data di bawah ini buatlah visualisasi seperti expected output:

```python

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(2*-np.pi, 2*np.pi, 200)

tan = np.tan(x)/10
cos = np.cos(x)
sin = np.sin(x)
```

```python
#code here
fig, axes = plt.subplots(1, 3, figsize=(10, 3))

axes[0].plot(x, tan, color='blue', label='tanX')
axes[0].set_title('Grafik tanx')
axes[0].legend()

axes[1].plot(x, sin, color='red', label='sinX')
axes[1].set_title('Grafik sinx')
axes[1].legend()

axes[2].plot(x, cos, color='green', label='cosX')
axes[2].set_title('Grafik cosx')
axes[2].legend()

plt.tight_layout()
plt.show()
```

<h2>Soal 2: Nested Axis</h2>

Dengan data di bawah ini, buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 100)
y = np.cos(x)
y2 = np.cos(x**2)
y3 = np.cos(x**3)
y4 = np.cos(x**4)
y5 = np.cos(x**5)
```

```python
#code here

fig, ax_main = plt.subplots(figsize=(10, 6))
ax_main.plot(x, y, label='cos(x)', color='blue')
ax_main.set_facecolor("white")
ax_main.set_title("Grafik CosX", color='black')
ax_main.tick_params(colors='black')
ax_main.legend()

inset_positions = [(-5, 0.55), (3, 0.55), (-5, -0.85), (3, -0.85)]
inset_data = [(y5, 'yellow', "Grafik CosX^5"),
              (y4, 'black', "Grafik CosX^4"),
              (y3, 'red', "Grafik CosX^3"),
              (y2, 'green', "Grafik CosX^2"),

              ]

for (x_pos, y_pos), (data, color, title) in zip(inset_positions, inset_data):
    inset_ax = fig.add_axes([0.5 + x_pos/12, 0.5 + y_pos/2, 0.2, 0.2])
    inset_ax.plot(x, y, color='blue', label="cos(x)")
    inset_ax.plot(x, data, color=color, label=title)
    inset_ax.set_title(title, fontsize=8)
    inset_ax.tick_params(labelsize=6)

plt.show()
```
