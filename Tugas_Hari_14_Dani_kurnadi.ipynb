```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('vgsales.csv')
df.head()
```



---



---



<h2>Soal 1: Preparing Data</h2>

Manipulasi data tersebut dengan menggroupkan berdasarkan Genre, kemudian ambil rata2 penjualan untuk Global_Sales berdasarkan kategori Genre.

```python
#code here

avg_sales_per_genre = df.groupby('Genre')['Global_Sales'].mean().reset_index()

avg_sales_per_genre = avg_sales_per_genre.sort_values(by='Genre')

avg_sales_per_genre
```

<h2>Soal 2: Perbandingan Kuantitatif Pieplot - Part a whole using Pie plot</h2>

1. Visualisasikan gambar yang tadi kita manipulasi, untuk menghasilkan visualisasi seperti di bawah ini
2. Tuliskan apa insight yang bisa kamu dapat dari visualisasi tersebut

```python
#code here

labels = avg_sales_per_genre['Genre']
sizes = avg_sales_per_genre['Global_Sales']
colors = plt.cm.tab20.colors

plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=colors)
plt.axis('equal')

plt.show()
```

<h2>Soal 3: Preparing Data</h2>

Manipulasi data tersebut dengan menggroupkan berdasarkan Genre, kemudian ambil rata2 penjualan untuk setiap Region Kecuali Global_Sales berdasarkan kategori Genre.

```python
#code here

avg_sales_per_region = df.groupby('Genre')[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].mean().reset_index()

avg_sales_per_region.sort_values(by='Genre')
```

<h2>Soal 4: Perbandingan Kuantitatif Barplot - Grouping Visualisasi Dengan Barplot</h2>

1. Visualisasikan gambar yang tadi kita manipulasi, untuk menghasilkan visualisasi seperti di bawah ini
2. Tuliskan apa insight yang bisa kamu dapat dari visualisasi tersebut

```python
genres = avg_sales_per_region['Genre']
regions = ['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']
values = avg_sales_per_region[regions]


x = np.arange(len(genres))
width = 0.5


fig, ax = plt.subplots(figsize=(12, 6))

num_genres = len(genres)
x = np.arange(len(regions))
width = 0.08

for i, genre in enumerate(genres):
    ax.bar(x + i * width, avg_sales_per_region.iloc[i, 1:], width, label=genre)


ax.set_xlabel('Region Sales')
ax.set_ylabel('Mean Sales')
ax.set_title('Mean Sales Video Games By Genre')
ax.set_xticks(x + (num_genres - 1) * width / 2)
ax.set_xticklabels(regions)
ax.legend(title="Genres", bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```

**Insight :**

Berdasarkan hasil visualisasi dapat dilihat bahwa region north america (NA_Sales) menjadi region dengan penjualan tertinggi dihampir semua genre , genre platform dan shooter terlihat menjadi genre dengan penjualan terbanyak.

Region Europe masih didominasi oleh genre platform dan shooter , namun rata rata penjualan masih dibawah region north america.

Sementara di region jepang penjualan terbanyak didominasi oleh genre role playing , hal ini tentu berbeda dengan region NA dan region EU , angka penjualan di region jepang lebih rendah dibanding 2 region sebelumnya.

dan di region lain , angka penjualan terbilang merata dengan shooter masih menjadi genre dengan hasil penjualan terbanyak

**Rekomendasi :**

Fokus pada genre genre yang populer diberbagai region , misal di region jepang bisa lebih di fokuskan pada pengembangan game dengan genre role playing , sambil berinovasi pada genre yang tidak populer seperti genre adventure.
