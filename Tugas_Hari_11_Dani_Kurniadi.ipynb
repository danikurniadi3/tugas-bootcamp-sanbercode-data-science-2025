<h2>Soal 1: Pemahaman Component Matplotlib</h2>

- Jelaskan apa itu Figure
- Jelaskan apa itu Axis

Figure adalah wadah utama yang merepresentasikan seluruh area grafik, termasuk
semua elemen seperti axis, judul, legenda, dan label.
Axis adalah bagian spesifik di dalam figure yang berisi grafik atau plot data. Axis memiliki sistem koordinat yang ditentukan oleh sumbu x dan y, yang digunakan untuk memvisualisasikan data.

<h2>Soal 2: Membuat Component Figure dan Axis</h2>

- Buatlah Figure dan Axis kosong tanpa Data
- Plot data di bawah ini dengan 2 cara, yaitu dengan membuat figure dan axis secara explicit dan Implicit

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 20, 100)
y = np.cos(x/2)
```

```python
# membuat Figure dan Axis dengan Data Kosong
fig = plt.figure()

ax = fig.add_subplot()
plt.show()
```

```python
#eksplisit

fig = plt.figure()

ax = fig.add_subplot()
ax.plot(x, y)
plt.show()
```

```python
#implicit
plt.plot(x, y)


plt.show()
```

<h2>Soal 3: Memasukan multiple data ke dalam satu axis</h2>

- Plot ketiga data berikut ke dalam satu axis, dan gunakan seaborn style
- save data tersebut dengan nama file bebas (apa saja) dan tampilkan kembali gambar tersebut.

```python
import seaborn as sns

x = np.linspace(0, 20, 100)
y = np.cos(x/2)
y2 = np.sin(x)
y3 = np.sin(2*x)
```

```python
plt.style.use('seaborn-v0_8-darkgrid')

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='cos(x/2)', color='blue', linewidth=2)
plt.plot(x, y2, label='sin(x)', color='green', linewidth=2)
plt.plot(x, y3, label='sin(2x)', color='red',linewidth=2)

plt.title("Plot of cos(x/2), sin(x), and sin(2x)", fontsize=16)
plt.xlabel("x", fontsize=14)
plt.ylabel("y", fontsize=14)
plt.legend(fontsize=12)
plt.grid(True)
```
