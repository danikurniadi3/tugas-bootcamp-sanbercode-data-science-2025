<h2>Soal 2: Data Restructuring</h2>

Jelaskan itu pivoting dan melting data

**Pivoting** adalah proses mengubah data dari format panjang (long format) menjadi format lebar (wide format) dengan memutar nilai kategori tertentu menjadi kolom baru, sehingga data lebih mudah dianalisis dalam bentuk agregat. Sebaliknya, **Melting** adalah proses mengubah data dari format lebar ke panjang untuk membuatnya lebih terstruktur . Pivoting digunakan untuk mengelompokkan data berdasarkan kategori, sementara melting digunakan untuk merapikan data dengan menggabungkan kolom menjadi pasangan atribut dan nilai, sehingga lebih fleksibel untuk analisis atau visualisasi.

<h2>Soal 3: Reshaping Data 1</h2>

```python
import pandas as pd
import random

data = {
    'Tipe Ujian' : ['UTS1', 'UAS1', 'UTS2', 'UAS2']*3,
    'Mata Pelajaran' : ['Fisika']*4 + ['Matematika']*4+['Statistika']*4,
    'Rata2 Waktu Belajar (menit)' : [random.choice(range(200, 800)) for i in range(12)],
    'Rata2 Nilai Ujian' : [random.choice(range(40, 100)) for i in range(12)]
}

df = pd.DataFrame(data)
df
```

Perhatikan data di atas, kemudian ubahlah bentuknya agar sesuai yang di harapkan.

```python
#code here
reshaped_df = df.pivot(index='Tipe Ujian', columns='Mata Pelajaran', values=['Rata2 Nilai Ujian', 'Rata2 Waktu Belajar (menit)'])

reshaped_df
```

<h2>Soal 4: Reshaping Data 2</h2>

```python
import pandas as pd

data = {
    'Hari' : ['Sabtu', 'Minggu'],
    'Kebun Binatang' : [271, 399],
    'Dufan' : [501, 700],
    'Disney' : [1000, 1001],
    'Bali' : [900, 803]
}
df = pd.DataFrame(data).set_index('Hari')
df
```

diatas adalah data tentang jumlah pengunjung di berbagai tempat hiburan, ubahlah struktur data tersebut sehingga sesuai dengan hasil yang diharapkan.

```python
#code here

reshaped_df = df.reset_index().melt(id_vars='Hari', var_name='Tempat Hiburan', value_name='Pengunjung')
reshaped_df
```

<h2>Soal 5: Data Enrichment</h2>

Lakukan binning data menggunakan qcut sehingga sesuai hasil yang diharapkan

```python
#code here

data = {
    'Tipe Ujian' : ['UTS1', 'UAS1', 'UTS2', 'UAS2']*3,
    'Mata Pelajaran' : ['Fisika']*4 + ['Matematika']*4+['Statistika']*4,
    'Rata2 Waktu Belajar (menit)' : [random.choice(range(200, 800)) for i in range(12)],
    'Rata2 Nilai Ujian' : [random.choice(range(40, 100)) for i in range(12)]
}

df = pd.DataFrame(data)

df['Kategori Kelulusan'] = pd.qcut(df['Rata2 Nilai Ujian'], q=3, labels=['Tidak Lulus', 'Remedial', 'Lulus'])
df.sort_values('Rata2 Nilai Ujian', ascending=False)

```
