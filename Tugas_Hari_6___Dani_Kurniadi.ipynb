<h2>Soal 3: DataFrame vs Series</h2>

- Apa perbedaan Series dengan list dan dictionary
- Apa perbedaan dataframe dan series
- Buatlah suatu data series

Jawab:



*   Series, list, dan dictionary adalah struktur data yang berbeda. List adalah kumpulan data yang terurut dan dapat diubah, sedangkan dictionary adalah kumpulan pasangan kunci-nilai. Series adalah struktur data satu dimensi yang merupakan bagian dari library Pandas, yang memungkinkan manipulasi dan analisis data yang efisien. Series memiliki indeks yang dapat disesuaikan dan mendukung operasi aritmatika dan statistik. Berbeda dengan list, series memiliki kemampuan untuk melakukan operasi secara vektoris dan mendukung data waktu dan tanggal.

*   DataFrame dan Series adalah dua struktur data utama dalam library Pandas. Series adalah struktur data satu dimensi yang menyerupai kolom tunggal dalam tabel, sedangkan DataFrame adalah struktur data dua dimensi yang menyerupai tabel dengan baris dan kolom. DataFrame terdiri dari kumpulan Series yang terkait dengan indeks yang sama. DataFrame mendukung operasi seperti penggabungan, pengurutan, dan pengelompokan data, sedangkan Series lebih cocok untuk operasi statistik dan analisisdatatunggal.

<h2>Soal 5: Deskriptif Statistik</h2>

Jelaskan apa itu deskriptis statistik!

Jawab:

Deskriptif statistik adalah metode analisis data yang berfokus pada pengumpulan, pengolahan dan penyajian data secara numerik dan grafis untuk memahami karakteristik data. Ini meliputi ukuran pusat (rata-rata, median, modus), ukuran variasi (rentang, variansi, simpangan baku), dan ukuran keragaman (koefisien variasi, indeks keragaman). Data disajikan dalam bentuk tabel, grafik (batang, lingkaran, garis), diagram, histogram dan grafik kotak-sayap. Analisis ini membantu mengidentifikasi pola, tren dan karakteristik data, sehingga memudahkan pengambilan keputusan dalam berbagai bidang seperti bisnis, kesehatan, pendidikandanpenelitian.

<h2>Soal 6: Practice Deskriptif Statistik</h2>

- Gunakan suatu metode dari pandas untuk mengetahui deskriptif statistik dari suatu data
- Berapa nilai rata-rata dari column price
- Berapa nilai standar deviasi dari column depth
- Berapa nilai maximum dari column carat

```python
# code here

df = pd.DataFrame(data)


stats = df.describe()
mean_price = stats.loc["mean", "price"]
std_depth = stats.loc["std", "depth"]
max_carat = stats.loc["max", "carat"]

mean_price, std_depth, max_carat
```

berikut adalah hasil analisis deskriptif statistik berdasarkan data pada dataset diamonds-1:
